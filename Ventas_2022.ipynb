```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
path = '/home/brandonealr/Desktop/ventas 2022/raiavl_mensual_venta_csv/conjunto_de_datos/raiavl_venta_mensual_tr_cifra_2022.csv'
data = pd.read_csv(path)
data = data.sort_values(by='UNI_VEH', ascending=False)
data
```

```python
top_20 = data.groupby('MARCA')[['UNI_VEH']].sum().sort_values(by='UNI_VEH', ascending=False).head(20)
print(top_20)
top_10 = data.groupby('MARCA')[['UNI_VEH']].sum().sort_values(by='UNI_VEH', ascending=False).head(10).sort_values(by='UNI_VEH', ascending=True)
```

```python
top_10.plot(kind='barh', color='mediumpurple').set_facecolor('aliceblue')
fig = plt.gcf()
fig.set_facecolor('aliceblue')
plt.xlabel('Autos vendidos', fontsize = 13)
plt.ylabel('Marca',fontsize = 13)
plt.title('Top 10 marcas con mayores ventas México 2022', color ='darkslateblue', fontsize = 15)
plt.xticks(rotation=90)
fig.show()
```

```python
segment = data.groupby('SEGMENTO')[['UNI_VEH']].sum().sort_values(by='UNI_VEH', ascending=False)
print(segment)
segment = data.groupby('SEGMENTO')[['UNI_VEH']].sum().sort_values(by='UNI_VEH', ascending=True)
```

```python
segment.plot(kind='barh', color='mediumpurple').set_facecolor('aliceblue')
fig = plt.gcf()
fig.set_facecolor('aliceblue')
plt.xlabel('Autos vendidos', fontsize = 13)
plt.ylabel('Segmento',fontsize = 13)
plt.title('Ventas de autos por segmento México 2022', color ='darkslateblue', fontsize = 15)
plt.xticks(rotation=90)
fig.show()
```

```python
total_vehicles = data['UNI_VEH'].sum()
segment['percentage'] = segment['UNI_VEH'] / total_vehicles * 100
segment.plot.pie(y='percentage', labels= segment.index ,autopct='%1.1f%%')
fig = plt.gcf()
fig.set_facecolor('aliceblue')
plt.legend(loc='upper left', bbox_to_anchor=(1.0, 0.5))
plt.ylabel('', fontsize = 13)
plt.xlabel('Porcentaje de Autos vendidos', fontsize = 13)
plt.title('Ventas de autos por segmento México 2022', color ='darkslateblue', fontsize = 15)
plt.xticks(rotation=90)
fig.show()
print(segment)
```

```python
monts = data.groupby('ID_MES')[['UNI_VEH']].sum().sort_values(by='ID_MES', ascending=False)
print(monts)
monts = data.groupby('ID_MES')[['UNI_VEH']].sum().sort_values(by='ID_MES', ascending=True)
```

```python
monts.plot(color='mediumpurple').set_facecolor('aliceblue')
fig = plt.gcf()
fig.set_facecolor('aliceblue')
plt.xlabel('Mes', fontsize = 13)
plt.ylabel('Autos vendidos',fontsize = 13)
plt.title('Ventas de autos por mes México 2022', color ='darkslateblue', fontsize = 15)
plt.xticks(rotation=90)
fig.show()
```

```python
origen = data.groupby('ORIGEN')[['UNI_VEH']].sum().sort_values(by='UNI_VEH', ascending=False)
print(origen)
origen = data.groupby('ORIGEN')[['UNI_VEH']].sum().sort_values(by='UNI_VEH', ascending=False).sort_values(by='UNI_VEH', ascending=True)
```

```python
origen['percentage'] = origen['UNI_VEH'] / total_vehicles * 100
origen.plot.pie(y='percentage', labels= origen.index ,autopct='%1.1f%%', colors=['Lightgreen', 'darkslateblue'])
fig = plt.gcf()
fig.set_facecolor('aliceblue')
plt.legend(loc='upper left', bbox_to_anchor=(1.0, 0.5))
plt.ylabel('', fontsize = 13)
plt.xlabel('', fontsize = 13)
plt.title('Ventas de autos por origen México 2022', color ='darkslateblue', fontsize = 15)
plt.xticks(rotation=90)
fig.show()
print(origen)
```
